```python
df=pd.read_csv('./listings.csv')
df.head()
```

```python
df=df.set_index(['cancellation_policy','review_scores_value'])
for group,frame in df.groupby(level=(0,1)):
    print(group)
```

```python
def grouping_fun(item):
    if item[1]=="10.0":
        return (item[0],"10.0")
    else:
        return (item[0],"not 10.0")

for group,frame in df.groupby(by=grouping_fun):
    print(group)
```

# Aggregation


```python
df=df.reset_index()
df.groupby("cancellation_policy").agg({"review_scores_value":np.average})
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value":np.nanmean})
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value":(np.nanmean,np.nanstd),"reviews_per_month":np.nanmean})
```

# Transformation


```python
cols=['cancellation_policy','review_scores_value']
tras_df=df[cols].groupby('cancellation_policy').transform(np.nanmean)
tras_df.head()
```

```python
tras_df.rename({'review_scores_value':'mean_review_scores'}, axis='columns', inplace=True)
df=df.merge(tras_df,left_index=True,right_index=True)
df.head()
```

```python
df['mean_diff']=np.absolute(df['review_scores_value']-df['mean_review_scores'])
df['mean_diff'].head()
```

# Filtering

```python
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
```

# Applying

```python
df=pd.read_csv("./listings.csv")
# And lets just include some of the columns we were interested in previously
df=df[['cancellation_policy','review_scores_value']]
df.head()
```

```python
def calc_mean_review_score(group):
    avg=np.nanmean(group['review_scores_value'])
    group["review_score_mean"]=np.abs(avg-group["review_scores_value"])
    return group
```

```python
df.groupby('cancellation_policy').apply(calc_mean_review_score).head()
```

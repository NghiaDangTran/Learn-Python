```python
df2=pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
def test_collumns(alpha=0.1):
    num_diff=0
    for col in df1.columns:
        teststat,pval=ttest_ind(df1[col],df2[col])
        if pval<=alpha:
            print(col,alpha,pval)
            num_diff+=1
    print(num_diff)
    
test_collumns(0.1)
```
